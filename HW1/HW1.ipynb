```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

cols = ['Top10perc', 'Apps', 'Enroll']

scatter_matrix(college[cols], figsize=(8,8))
plt.show()
```

```python
college.boxplot(column='Outstate', by='Private', figsize=(6,6))
plt.title('Outstate Tuition by Private/Public')
plt.suptitle('')  
plt.ylabel('Outstate Tuition')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

college['Elite'] = pd.cut(college['Top10perc'],
                          bins=[0,50,100],
                          labels=['No','Yes'])

print(college['Elite'].value_counts())

college.boxplot(column='Outstate', by='Elite', figsize=(6,6))
plt.title('Outstate Tuition by Elite Status')
plt.suptitle('')  
plt.ylabel('Outstate Tuition')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

college['Apps'].plot.hist(bins=20, ax=axes[0,0], title='Applications')
college['Accept'].plot.hist(bins=20, ax=axes[0,1], title='Accepted')
college['Enroll'].plot.hist(bins=15, ax=axes[1,0], title='Enrolled')
college['Outstate'].plot.hist(bins=25, ax=axes[1,1], title='Out-of-state Tuition')

plt.tight_layout() 
plt.show()
```
