```python
# (c)
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

distances = np.array([0.3, 0.4, 0.7, 0.5, 0.8, 0.45])

# Complete linkage clustering
Z_complete = sch.linkage(distances, method='complete')

# dendrogram
plt.figure(figsize=(8, 5))
dendro = sch.dendrogram(Z_complete, labels=["Obs1","Obs2","Obs3","Obs4"])

# cutting position
cut_height = 0.5
plt.axhline(y=cut_height, c='red', linestyle='--', label=f"cut at height {cut_height}")

plt.title("Complete Linkage Dendrogram with Cut")
plt.ylabel("Distance (Height)")
plt.legend()
plt.show()

from scipy.cluster.hierarchy import fcluster
clusters = fcluster(Z_complete, t=cut_height, criterion='distance')

print("Clusters after cutting dendrogram at height =", cut_height)
for i, cluster_id in enumerate(clusters, start=1):
    print(f"Obs{i}: Cluster {cluster_id}")
```

```python
# (d)
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster
distances = np.array([0.3, 0.4, 0.7, 0.5, 0.8, 0.45])

# Perform Single Linkage hierarchical clustering
Z_single = sch.linkage(distances, method='single')

# dendrogram
plt.figure(figsize=(8, 5))
dendro = sch.dendrogram(Z_single, labels=["Obs1","Obs2","Obs3","Obs4"])

# Choose a height between second and third merge heights (from linkage matrix)
cut_height = 0.42
plt.axhline(y=cut_height, color='red', linestyle='--', label=f"Cut at height {cut_height}")

plt.title("Single Linkage Dendrogram with Cut")
plt.ylabel("Distance (Height)")
plt.legend()
plt.show()

# Extract cluster assignments 
clusters = fcluster(Z_single, t=cut_height, criterion='distance')

print("Clusters after cutting dendrogram at height =", cut_height)
for i, cluster_id in enumerate(clusters, start=1):
    print(f"Obs{i}: Cluster {cluster_id}")
```

```python
# (e)
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

distances = np.array([0.3, 0.4, 0.7, 0.5, 0.8, 0.45])
Z_complete = sch.linkage(distances, method='complete')

plt.figure(figsize=(8, 5))
# Change leaf rotation to swap positions (equivalent dendrogram)
sch.dendrogram(Z_complete, labels=["Obs3","Obs4","Obs1","Obs2"], leaf_rotation=90)
plt.title("Equivalent Dendrogram (Complete Linkage)")
plt.ylabel("Distance (Height)")
plt.show()
```
